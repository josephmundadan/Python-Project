```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
```

```python
df=pd.read_csv('paython//myexcel - myexcel.csv.csv')
df.head()
df.info()
```

```python
df['Height']=np.random.randint(150,180,size=len(df))
df.head()
```

```python
#1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)
```

```python
team_distribution=df['Team'].value_counts()
team_distribution
```

```python
team_percentage = (team_distribution / len(df)) * 100
team_percentage
```

```python
plt.figure(figsize=(8,5),dpi=100)
plt.xlabel('team distribution')
plt.ylabel('team_percentage')
plt.title('Bar Chart for Employee Ration')
sns.barplot(x=team_distribution.index,y=team_percentage.values,color='red')
plt.xticks(rotation=90)
plt.tight_layout()


        
```

```python
#2. Segregate employees based on their positions within the company. (2 marks)
```

```python
Segregate_employees=df['Position'].value_counts()
Segregate_employees
```

```python
plt.figure(figsize=(8,5),dpi=500)
plt.xlabel('Employees')
plt.ylabel('Segregate_employees')
plt.title('bar char for "Segregate employees"')
plt.tight_layout()
sns.barplot(x=Segregate_employees.index,y=Segregate_employees.values,color='red')
```

```python
# 3. Identify the predominant age group among employees. (2 marks)
```

```python
bins=[0,20,25,30,35,40,45,50]
labels=['<20','20-25','25-30','30-35','35-40','40-45','45+']
df['Age Group']=pd.cut(df['Age'],bins=bins,labels=labels,right=False)
df
age_group= df['Age Group'].value_counts().sort_index()
age_group
```

```python
plt.figure(figsize=(8,5),dpi=300)
plt.xlabel('age range')
plt.ylabel('number of employees')
plt.title('bar chart')
sns.barplot(x=age_group.index,y=age_group.values,color='r')
plt.tight_layout()
plt.show()
```

```python
4. Discover which team and position have the highest salary expenditure. (2 marks)
```

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5),dpi=300)
plt.xlabel('team')
plt.ylabel('salary')
plt.title('team salary')
sns.barplot(x=team_salary.index,y=team_salary.values)
plt.xticks(rotation=90)
plt.yticks([30000000,60000000,70000000,80000000,90000000,100000000])
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,5),dpi=300)
plt.xlabel('team')
plt.ylabel('salary')
plt.title('team salary')
sns.barplot(x=position_salary.index,y=position_salary.values)
plt.tight_layout()
```

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
plt.figure(figsize=(5,8),dpi=300)
plt.xlabel('team')
plt.ylabel('salary')
plt.title('team salary')
sns.barplot(x=team_salary.index,y=team_salary.values)
plt.xticks(rotation=90)
plt.yticks([30000000,60000000,70000000,80000000,90000000,100000000])
plt.tight_layout()
```

```python
#5. Investigate if there's any correlation between age and salary, and represent it visually. (2 marks)
```

```python
# 5. Correlation Between Age and Salary
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Age'], y=df['Salary'],data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary ($)')
plt.tight_layout()
plt.show()

```

```python
correlation=df[['Salary','Age']].corr()
correlation
corr=correlation.iloc[0,1]
corr
print(f"the correlation between age and salary {corr:0.4f}")
```
